```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats
```

```python
colors = 'red'

mu = [0] # mean
sd = 1 # standard deviation

X = np.arange(-4,4,0.1)
Y = scipy.stats.norm.pdf(X, loc=mu, scale=sd)
```

```python
fig = plt.figure(figsize=(10, 4), facecolor="white")

tick = [i for i in range(-4, 4, 1)]

ax = plt.subplot()
ax.set_xticks(tick)

ax.plot(X, Y, c=colors, zorder=10)
ax.grid(axis='x', c='lightgray', zorder=9)
ax.grid(axis='y', c='lightgray', zorder=9)
[ax.spines[side].set_visible(False) for side in ['right','top']]

plt.show()
```
